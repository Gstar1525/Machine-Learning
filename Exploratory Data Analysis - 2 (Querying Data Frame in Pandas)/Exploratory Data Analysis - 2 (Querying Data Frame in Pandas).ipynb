**Task** - Sorting using sort_values() function

```python
df.sort_values(by=['Age'], ascending=[True])
```

**Task** - Mapping the values with new values using map() and replace().

```python
new={'male':0,'female':1}
df['Sex']=df['Sex'].map(new)
df.head()
```

```python
old={0:'Died',1:'Survived'}
df=df.replace({'Survived':old})
df.head()
```

**Task** - Grouping values using groupby().

```python
df.groupby(by='Survived')['Age'].describe()
```

**Task** - Relating two or more than two variables using crosstab().

```python
pd.crosstab(df['Sex'],df['Survived'], normalize=True)
```
